```python
from sklearn.datasets import load_breast_cancer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
import pandas as pd
import numpy as np

data_bc = load_breast_cancer()
df_bc = pd.DataFrame(data_bc['data'], columns=data_bc['feature_names'])
tabel = df_bc.head(6)
print(tabel)
x = data_bc['data']
print(x)
y = data_bc['target']
names = data_bc['target_names']
print(names)

#bagi data menjadi train set dan test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=10)

#buat definisi parameter grid yang ingin dicoba
param_grid = {'n_neighbors':np.arange(2,20), 'weights': ['distance', 'uniform'], 'p': [1,2], 'algorithm': ['auto', 'brute', 'kd_tree', 'ball_tree']}

#modeling dengan KNN + Randomized Search
model_knn = KNeighborsClassifier()
gscv = GridSearchCV(model_knn, param_grid, scoring='accuracy', cv=5)
model = gscv.fit(x_train, y_train)
best_params = gscv.best_params_
best_score = gscv.best_score_
print('\n')
print(model)
print('\n')
print(best_params)
print('\n')
print(best_score)
```

```python
from sklearn.datasets import load_boston
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import Normalizer
import pandas as pd

boston = load_boston()

x = boston['data']
y = boston['target']

model_lr = LinearRegression()
model_lr.fit(x, y)

print('Koefisien model linear regression: ')
print(model_lr.coef_)
print('\n')
print('intercept: ', model_lr.intercept_)
print('\n')

#normalisasi data
norm = Normalizer().fit_transform(x)
df_boston = pd.DataFrame(norm, columns=boston['feature_names'])
normalisasi = df_boston.head(6)
print(normalisasi)
print('\n')

#mengaplikasikan salah satu Dimensionality Redustion Algorithms yakni Principal Component Analysis
pca = PCA(n_components=2).fit_transform(norm)
df_pca = pd.DataFrame(data=pca, columns=['Fitur 1', 'Fitur 2'])
PCA = df_pca.head(6)
print(PCA)
```

```python
from sklearn import datasets
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA

iris = datasets.load_iris()
atribut = iris.data
label = iris.target

x_train, x_test, y_train, y_test = train_test_split(atribut, label, test_size=0.2)

#menghitung akurasi tanpa PCA, melainkan tree
decision_tree = tree.DecisionTreeClassifier()
model_pertama = decision_tree.fit(x_train, y_train)
akurasi_tree = model_pertama.score(x_test, y_test)
print(akurasi_tree)
print('\n')

#menghitung akurasi dengan PCA
pca = PCA(n_components=4) #buat objek PCA dengn 4 principal components/attributes
pca_attributes = pca.fit_transform(x_train)
variasi_pca = pca.explained_variance_ratio_
print(variasi_pca)
print('\n')

#PCA dengan 2 principal components terbaik
pca = PCA(n_components=2) #karena 0.96 sudah termasuk tinggi
x_train_pca = pca.fit_transform(x_train)
x_test_pca = pca.fit_transform(x_test)

#uji akurasi classifier setelah mengggunakan pca
model2 = decision_tree.fit(x_train_pca, y_train)
akurasi_pca = model2.score(x_test_pca, y_test)
print(akurasi_pca)
```
